```python
%matplotlib inline
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# create an empty graph
G = nx.Graph()
```

```python
type(G)
```

```python
print(G.nodes())
print(G.edges())
print(type(G.nodes()))
print(type(G.edges()))
```

```python
G.add_node('a')
G.add_nodes_from(['b', 'c'])
```

```python
print('Nodes of the graph G: ')
print(G.nodes())
print('Edges of the graph G: ')
print(G.edges())
```

```python
G.add_edge(1, 2)
edge = ('d', 'e')
G.add_edge(*edge)
edge = ('a', 'b')
G.add_edge(*edge)
```

```python
print('Nodes of the graph G: ')
print(G.nodes())
print('Edges of the graph G: ')
print(G.edges())
```

```python
G.add_edges_from([('a', 'c'), ('c', 'd'), 
                  ('a', 1), (1, 'd'),
                  ('a', 2)])
```

```python
print('Nodes of the graph G: ')
print(G.nodes())
print('Edges of the graph G: ')
print(G.edges())
```

# Draw Graph

```python
nx.draw(G)
plt.show()
```

```python
nx.draw(G, with_labels=True,
       font_weight='bold')
plt.show()
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

# Generate Path Graph

```python
G = nx.path_graph(4)

print('Nodes of the graph G: ')
print(G.nodes())
print('Edges of the graph G: ')
print(G.edges())
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

# Petersen Graph and Shell

```python
G = nx.petersen_graph()

print('Nodes of the graph G: ')
print(G.nodes())
print('Edges of the graph G: ')
print(G.edges())
```

```python
nx.draw(G, with_labels=True)
plt.show()
```

```python
nx.draw_shell(G, nlist=[range(5, 10),
                        range(5)],
              with_labels=True,
              font_weight='bold')
plt.show()
```

```python
G = nx.path_graph(4)
cities = {0: 'Mumbai', 1: 'Hyderabad',
          2: 'Banglore', 3: 'New York'}
H=nx.relabel_nodes(G, cities)

print('Nodes of the graph H: ')
print(H.nodes())
print('Edges of the graph H: ')
print(H.edges())
```

```python
nx.draw(H, with_labels=True)
plt.show()
```

```python
options = {'node_color': 'black',
           'node_size': 100,
           'width': 3}
```

```python
G = nx.petersen_graph()
nx.draw_random(G, **options)
plt.show()
```

```python
nx.draw_circular(G, **options)
plt.show()
```

```python
nx.draw_spectral(G, **options)
plt.show()
```

```python
nx.draw_shell(G, nlist=[range(5, 10),
                        range(5)],
              **options)
plt.show()
```

```python
G = nx.dodecahedral_graph()
shells = [[2, 3, 4, 5, 6],
          [8, 1, 0, 19, 18,
           17, 16, 15, 14, 7],
          [9, 10, 11, 12, 13]]
nx.draw_shell(G, nlist=shells,
              **options)
plt.show()
plt.savefig('graph.png')
```

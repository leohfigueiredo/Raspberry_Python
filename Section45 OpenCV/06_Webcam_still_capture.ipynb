```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture(0)
```

```python
if cap.isOpened():
    ret, frame = cap.read()
    print(ret)
    print(frame)
else:
    ret = False
```

```python
print(type(frame))
```

```python
img1 = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
plt.title('Webcam snapshot')
plt.show()

cap.release()
```

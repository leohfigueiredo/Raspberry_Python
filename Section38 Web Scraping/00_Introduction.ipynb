```python
import requests

req = requests.get('https://linux.org//')
```

```python
print(req.encoding)
print(req.status_code)
print(req.elapsed)
print(req.url)
print(req.history)
print(req.headers['Content-Type'])
```

```python
from bs4 import BeautifulSoup

data = req.text
soup = BeautifulSoup(data)

for link in soup.find_all('a'):
    print(link.get('href'))
```
